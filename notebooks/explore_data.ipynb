### Customer Data Issues:
- we will be keeping only customer_id as cust_id is nothing but serial no and it will be primary key
- we will be keeping only full_name and merging with customer_name and removing all email address from there
- we will be changing phone colum name to alternate_phone , as phone_number colum has 0 null values
- we will be removing registration_date colum as it has 96 null values and keeping reg_date colum as it has 0 null values
- we will be merging email and email_address colums
- we will be merging status and customer_status and values will be in lower case in this colum and keeping only status colum as it has less null values
- Also Customer Gender some where is M Some where completely written Male some where F some where completely written Female some where O some where completely written Other so we will only use M , F ,O 

```python
import pandas as pd

# Load raw JSON
customers = pd.read_json('../data/customers_messy_data.json')

# Step 1: Drop cust_id
customers.drop(columns=['cust_id'], inplace=True)

# Step 2: Merge names (assuming full_name is corrupted by emails)
customers['customer_name'] = customers['customer_name'].combine_first(customers['full_name'])
customers.drop(columns=['full_name'], inplace=True)

# Step 3: Handle phone numbers
customers.rename(columns={'phone': 'alternate_phone'}, inplace=True)

# Step 4: Handle registration dates
customers['registration_date'] = customers['reg_date']
customers.drop(columns=['reg_date'], inplace=True)

# Step 5: Merge emails
customers['email'] = customers['email'].combine_first(customers['email_address'])
customers.drop(columns=['email_address'], inplace=True)

# Step 6: Merge status
customers['status'] = customers['status'].combine_first(customers['customer_status'])
customers['status'] = customers['status'].str.lower()
customers.drop(columns=['customer_status'], inplace=True)

# Step 7: Clean gender values (standardize to M / F / O)
customers['gender'] = customers['gender'].astype(str).str.strip().str.lower().map({
    'm': 'M', 'male': 'M',
    'f': 'F', 'female': 'F',
    'o': 'O', 'other': 'O'
})
```

```python
import numpy as np

is_email_in_name = customers['customer_name'].str.contains('@', na=False)

is_email_empty = customers['email'].isna() | (customers['email'] == '')

customers.loc[is_email_in_name & is_email_empty, 'email'] = customers.loc[is_email_in_name & is_email_empty, 'customer_name']

customers.loc[is_email_in_name, 'customer_name'] = np.nan
```

### Orders Data Issues & Cleaning Plan

1. **Primary Key:**
   - Keep only `order_id` as primary key.
   - Drop `ord_id` (just a serial number).

2. **Customer Reference:**
   - Use `cust_id` instead of `customer_id` for consistency and professionalism.
   - Drop `customer_id`.

3. **Order Date Columns:**
   - Keep only `order_date`.
   - Where `order_date` is null, fill with `order_datetime`.
   - Drop `order_datetime` after merge.

4. **Date Format:**
   - Normalize all dates to `DD-MM-YYYY` format, regardless of original format (`MM/DD/YYYY`, `YYYY-MM-DD`, etc.)

5. **Related Fields (Uncertain):**
   - Keep both `product_id` and `item_id` since values differ.
   - Keep both `quantity` and `qty` since values differ.
   - Keep both `unit_price` and `price` since values differ.
   - Keep both `total_amount` and `order_total` since values differ.

6. **Status Columns:**
   - Merge `status` and `order_status`:
     - Keep only `status`
     - If `status` is null, use `order_status`
     - Convert to lowercase for consistency
     - Drop `order_status` column


```python
import pandas as pd
import numpy as np
from datetime import datetime

# Load data
orders = pd.read_csv("../data/orders_unstructured_data.csv")

# Drop ord_id and customer_id
orders.drop(columns=['ord_id', 'customer_id'], inplace=True)

# Fill missing order_date with order_datetime
orders['order_date'] = orders['order_date'].fillna(orders['order_datetime'])

# Drop order_datetime
orders.drop(columns=['order_datetime'], inplace=True)

# Step 3: Standardize all dates to DD-MM-YYYY
def parse_date(date_str):
    if pd.isnull(date_str):
        return np.nan
    for fmt in ("%d-%m-%Y", "%d-%m-%Y %H:%M:%S", "%m/%d/%Y", "%Y-%m-%d", "%Y-%m-%d %H:%M:%S"):
        try:
            return pd.to_datetime(date_str, format=fmt).strftime("%d-%m-%Y")
        except:
            continue
    return np.nan

orders['order_date'] = orders['order_date'].apply(parse_date)

# Step 4: Merge status and order_status
orders['status'] = orders['status'].combine_first(orders['order_status'])
orders['status'] = orders['status'].str.lower()
orders.drop(columns=['order_status'], inplace=True)

# Optional: Reset index and preview
orders.reset_index(drop=True, inplace=True)
orders.head()
```

```python
orders.to_csv("../cleaned_data/cleaned_orders.csv", index=False)
```

### Product Data Issues & Cleaning Plan

1. **Primary Key:**
   - Keep only `product_id` as primary key.
   - Drop `item_id` (just a serial number).

2. **Product Name Columns:**
   - Keep only `product_name`, drop `item_name` (assumed duplicate).

3. **Category Columns:**
   - Keep only `category`
   - Drop `product_category`
   - Even though values differ, they represent the same concept and both have 0 nulls.

4. **Remaining Columns:**
   - Rest of the fields (`description`, `brand`, `price`, `stock_quantity`, etc.) look consistent and usable.


```python
# Load product data
products = pd.read_json("../data/products_inconsistent_data.json")

# Drop 'item_id' and 'item_name'
products.drop(columns=['item_id', 'item_name'], inplace=True)

# Drop 'product_category' (keep only 'category')
products.drop(columns=['product_category'], inplace=True)

# Optional: Clean string columns like 'is_active' to booleans if needed
products['is_active'] = products['is_active'].astype(str).str.lower().map({
    'yes': True, 'true': True, '1': True,
    'no': False, 'false': False, '0': False
})

# Reset index
products.reset_index(drop=True, inplace=True)

# Preview cleaned data
products.head()
```

```python
products.to_csv("../cleaned_data/cleaned_products.csv", index=False)
```
